# conditional Selection

```python
data= {'A': [1, 2, 3, 4, 5], 'B': ['X', 'Y', 'X', 'Y', 'Z']}
df = pd.DataFrame(data)
df
```

```python
# Select rows where column 'A' is greater than 2
result = df[df['A'] > 2]
result
```

```python
result=df[df['B']=='X']
result
```

```python
result = df[(df['A'] > 2) & (df['B'] == 'Y')]
result
```

```python
result = df[(df['A'] > 0) & (df['B'] == 'Y')]
result
```

```python
# Select rows where column 'A' is not equal to 2
result = df[~(df['A'] == 2)]
result
```
